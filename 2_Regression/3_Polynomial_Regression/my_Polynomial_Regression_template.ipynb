# Polynomial Regression Template

## ======== Import the data set ========

```python
import pandas as pd
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, -1].values
```

## ========= How linear regression model does? ====

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

### ========= Visualize the simple linear regression result =======

```python
import matplotlib.pyplot as plt
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg.predict(x), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show() #does not look like a simple linear model.
```

## ======== Try polynomial regression model ========

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 5)
x_poly = poly_reg.fit_transform(x)

lin_reg_poly = LinearRegression()
lin_reg_poly.fit(x_poly, y)
```

### ========= Visualize the polynomial regression result =======

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg_poly.predict(x_poly), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

### ========= finer mesh visualization =======

```python
import numpy as np
x_grid = np.arange(min(x), max(x)+0.1, 0.1).reshape(-1, 1)
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, lin_reg_poly.predict(poly_reg.fit_transform(x_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## ======== Predict a single observation ========

```python
print ('Simple Linear Regression predicts', lin_reg.predict(np.array([6.5]).reshape((1, -1)))[0])
print ('Polynomial Regression predicts', lin_reg_poly.predict(poly_reg.fit_transform(np.array([6.5]).reshape((1, -1))))[0])
```
