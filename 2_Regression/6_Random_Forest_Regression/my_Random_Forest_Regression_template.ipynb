# Random Forest Regression template

## ======== Import the data set ========

```python
import pandas as pd
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, -1].values
```

## ====== Fitting random forest regression to data and visualize it ======

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import matplotlib.pyplot as plt

for n_trees in [2, 5, 10]:
    regressor = RandomForestRegressor(n_estimators=n_trees, random_state=0)
    regressor.fit(x, y)
    ##visualize
    x_grid = np.arange(min(x), max(x)+0.01, 0.01).reshape(-1, 1)
    plt.scatter(x, y, color = 'red')
    plt.plot(x_grid, regressor.predict(x_grid), color = 'blue')
    plt.title('Truth or Bluff (Random Forest ('+str(n_trees)+' trees) Regression)')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()
```

## ======= Predict a single observation =======

```python
print ('At 6.5, the prediction is', regressor.predict(np.array([6.5]).reshape((1,1)))[0])
```
